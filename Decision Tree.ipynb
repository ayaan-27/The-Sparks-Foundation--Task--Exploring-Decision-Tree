# Task Title: Exploring the Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

## Loading the data 

```python
data= load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
print(df.head(10))
target=data.target
```

```python
df.describe()
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df,target,test_size=0.3,random_state=10)
```

## Training the classifier

```python
classifier=DecisionTreeClassifier(random_state=12)
classifier.fit(X_train,Y_train)
```

## Visualising the decision tree

```python
figure=plt.figure(figsize=(15,10))
DT=tree.plot_tree(classifier,feature_names=data.feature_names,class_names=data.target_names,filled=True)
```
